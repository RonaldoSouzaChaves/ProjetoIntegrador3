Limpeza e tratamento da tabela TipoCrime.

```python
#chamando biblioteca
import pandas as pd

#lendo csv
tipoCrime = pd.read_csv('C:/Users/PICHAU/Downloads/iucr_lista.csv', sep=';', encoding='utf-8')

#criando dataframe
tipoCrime = pd.DataFrame(tipoCrime)

tipoCrime
```

```python
#apagando colunas desnecessárias
del tipoCrime['ACTIVE']
del tipoCrime['INDEX CODE']
tipoCrime
```

```python
#apagando linhas com valores nulos na coluna FBI_code
tipoCrime.dropna(subset=['FBI_code'], inplace=True)

tipoCrime
```

```python
#renomeando as colunas
tipoCrime.rename(columns=
 {"PRIMARY DESCRIPTION": "CategoriaCrime",
  "SECONDARY DESCRIPTION": "DescricaoCrime",
  "FBI_code": "CodigoFBI"},
inplace=True)

tipoCrime
```

```python
#gerando csv
tipoCrime.to_csv('tipoCrime.csv')
```
