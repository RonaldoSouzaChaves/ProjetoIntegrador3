```python
# importando bibliotecas
import pandas as pd
```

```python
# importanto as bases de dados
casos = pd.read_csv('C:/Users/PICHAU/Downloads/Crimes_-_2001_to_Present_20231016.csv', sep=',', encoding='utf-8', on_bad_lines='skip')

casos
```

```python
#apagando colunas desnecessárias
del casos['Block']
del casos['Primary Type']
del casos['Description']
del casos['District']
del casos['FBI Code']
del casos['X Coordinate']
del casos['Y Coordinate']
del casos['Year']
casos
```

```python
#renomeando as colunas
casos.rename(columns=
 {"Case Number": "NumeroCaso",
  "Date": "Data",
  "Location Description": "DescricaoLocal",
  "Arrest": "Prisao",
  "Domestic": "Domestico",
  "Ward": "Ala",
  "Community Area" : "AreaComunitaria",
  "Updated On": "UltimaAtualizacao",
  "Location": "Local"},
inplace=True)
```

```python
# Verificando valores null
print(casos['NumeroCaso'].isna().sum().sum()) # 7 
print(casos['DescricaoLocal'].isna().sum().sum()) # 1990
print(casos['Latitude'].isna().sum().sum()) # 105573
print(casos['Longitude'].isna().sum().sum()) # 105574
print(casos['Local'].isna().sum().sum()) # 105574
```

```python
##Removendo null
casos = casos.dropna(subset=['Local'])
casos = casos.dropna(subset=['NumeroCaso'])
casos = casos.dropna(subset=['DescricaoLocal'])
```

```python
# Verificando o tipo 
casos.dtypes
```

```python
# Transformando tipo de variavel 
casos['NumeroCaso'] = casos['NumeroCaso'].astype(str)
casos['IUCR'] = casos['IUCR'].astype(str)
casos['DescricaoLocal'] = casos['DescricaoLocal'].astype(str)
casos['Latitude'] = casos['Latitude'].astype(float)
```

```python
#Conversão da data e tempo
# Data
# Converta a coluna "data_hora" de "object" para "datetime"
casos['Data'] = pd.to_datetime(casos['Data'])

# Converta datetime para uma string formatada
casos['Data'] = casos['Data'].dt.strftime('%Y-%m-%d %H:%M:%S')


# Ultima Atulização data 
# Converta a coluna "data_hora" de "object" para "datetime"
casos['UltimaAtualizacao'] = pd.to_datetime(casos['UltimaAtualizacao'])

# # Converta datetime para uma string formatada
casos['UltimaAtualizacao'] = casos['UltimaAtualizacao'].dt.strftime('%Y-%m-%d %H:%M:%S')
```

```python
#gerando csv
casos.to_csv('casos.csv')
```
